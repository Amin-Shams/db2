```python
# import pandas as pd
# import numpy as np
# import random
# from datetime import datetime, timedelta
# from faker import Faker

# fake = Faker()

# num_invoice_lines = 1_000_000
# num_payments = 150_000
# num_recognitions = 200_000

# num_customers = 500
# num_contracts = 1000
# num_service_types = 100
# num_taxes = 50
# num_tariffs = 200
# num_invoices = 10000
# num_billing_cycles = 5

# # تولید Customer
# customers = pd.DataFrame({
#     "CustomerID": range(1, num_customers + 1),
#     "CustomerCode": [f"CUST{str(i).zfill(4)}" for i in range(1, num_customers + 1)],
#     "CustomerName": [fake.company() for _ in range(num_customers)],
#     "CustomerType": np.random.choice(['Individual', 'Company', 'Foreign'], num_customers),
#     "TIN": [fake.uuid4()[:10] for _ in range(num_customers)],
#     "VATNumber": [fake.uuid4()[:10] for _ in range(num_customers)],
#     "Phone": [fake.phone_number() for _ in range(num_customers)],
#     "Email": [fake.email() for _ in range(num_customers)],
#     "Address": [fake.address().replace("\n", ", ") for _ in range(num_customers)],
#     "CountryID": np.random.randint(1, 10, num_customers)
# })

# # BillingCycle
# billing_cycles = pd.DataFrame({
#     "BillingCycleID": range(1, num_billing_cycles + 1),
#     "CycleName": ['Monthly', 'Quarterly', 'Annually', 'Bi-Weekly', 'Weekly'],
#     "CycleLengthInDays": [30, 90, 365, 14, 7]
# })

# # Contracts
# contracts = pd.DataFrame({
#     "ContractID": range(1, num_contracts + 1),
#     "CustomerID": np.random.choice(customers["CustomerID"], num_contracts),
#     "ContractNumber": [f"CON{str(i).zfill(5)}" for i in range(1, num_contracts + 1)],
#     "StartDate": [datetime(2022, 1, 1) + timedelta(days=random.randint(0, 500)) for _ in range(num_contracts)],
#     "EndDate": [datetime(2024, 1, 1) + timedelta(days=random.randint(0, 365)) for _ in range(num_contracts)],
#     "BillingCycleID": np.random.choice(billing_cycles["BillingCycleID"], num_contracts),
#     "PaymentTerms": np.random.choice(['Net 30', 'Net 60', 'Advance'], num_contracts),
#     "ContractStatus": np.random.choice(['Active', 'Expired'], num_contracts),
#     "CreatedDate": [datetime.now() - timedelta(days=random.randint(0, 730)) for _ in range(num_contracts)]
# })

# # ServiceType
# service_types = pd.DataFrame({
#     "ServiceTypeID": range(1, num_service_types + 1),
#     "ServiceName": [f"Service {i}" for i in range(1, num_service_types + 1)],
#     "ServiceCategory": np.random.choice(['Unloading', 'Storage', 'Transport'], num_service_types),
#     "BaseRate": np.round(np.random.uniform(100, 1000, num_service_types), 2),
#     "UnitOfMeasure": np.random.choice(['TEU', 'Hour', 'Ton'], num_service_types),
#     "Taxable": np.random.choice([0, 1], num_service_types),
#     "IsActive": 1
# })

# # Tax
# taxes = pd.DataFrame({
#     "TaxID": range(1, num_taxes + 1),
#     "TaxName": [f"Tax {i}" for i in range(1, num_taxes + 1)],
#     "TaxRate": np.round(np.random.uniform(0.01, 0.2, num_taxes), 2),
#     "TaxType": np.random.choice(['National', 'Service'], num_taxes),
#     "EffectiveFrom": [datetime(2020, 1, 1) + timedelta(days=i*5) for i in range(num_taxes)],
#     "EffectiveTo": [None]*num_taxes
# })

# # Tariff
# tariffs = pd.DataFrame({
#     "TariffID": range(1, num_tariffs + 1),
#     "ServiceTypeID": np.random.choice(service_types["ServiceTypeID"], num_tariffs),
#     "ValidFrom": [datetime(2021, 1, 1) + timedelta(days=random.randint(0, 1000)) for _ in range(num_tariffs)],
#     "ValidTo": [None]*num_tariffs,
#     "UnitRate": np.round(np.random.uniform(150, 1200, num_tariffs), 2)
# })

# # Invoices
# invoices = pd.DataFrame({
#     "InvoiceID": range(1, num_invoices + 1),
#     "ContractID": np.random.choice(contracts["ContractID"], num_invoices),
#     "InvoiceNumber": [f"INV{str(i).zfill(6)}" for i in range(1, num_invoices + 1)],
#     "InvoiceDate": [datetime(2024, 1, 1) + timedelta(days=random.randint(0, 180)) for _ in range(num_invoices)],
#     "DueDate": [datetime(2024, 7, 1) + timedelta(days=random.randint(0, 90)) for _ in range(num_invoices)],
#     "Status": np.random.choice(['Paid', 'Overdue', 'Cancelled'], num_invoices),
#     "TotalAmount": np.round(np.random.uniform(1000, 15000, num_invoices), 2),
#     "TaxAmount": np.round(np.random.uniform(100, 2000, num_invoices), 2),
#     "CreatedBy": [fake.user_name() for _ in range(num_invoices)],
#     "CreatedDate": [datetime.now() - timedelta(days=random.randint(0, 100)) for _ in range(num_invoices)]
# })

# invoice_line_df = pd.DataFrame({
#     "InvoiceLineID": range(1, num_invoice_lines + 1),
#     "InvoiceID": np.random.choice(invoices["InvoiceID"], num_invoice_lines),
#     "ServiceTypeID": np.random.choice(service_types["ServiceTypeID"], num_invoice_lines),
#     "TaxID": np.random.choice(taxes["TaxID"], num_invoice_lines),
#     "Quantity": np.random.randint(1, 20, num_invoice_lines),
#     "UnitPrice": np.round(np.random.uniform(100, 500, num_invoice_lines), 2),
#     "DiscountPercent": np.round(np.random.uniform(0, 0.2, num_invoice_lines), 2),
#     "TaxAmount": np.round(np.random.uniform(10, 150, num_invoice_lines), 2),
#     "NetAmount": np.round(np.random.uniform(100, 1000, num_invoice_lines), 2)
# })

# payment_df = pd.DataFrame({
#     "PaymentID": range(1, num_payments + 1),
#     "InvoiceID": np.random.choice(invoices["InvoiceID"], num_payments),
#     "PaymentDate": [datetime(2024, 1, 1) + timedelta(days=random.randint(0, 180)) for _ in range(num_payments)],
#     "Amount": np.round(np.random.uniform(100, 8000, num_payments), 2),
#     "PaymentMethod": np.random.choice(['Cash', 'Card', 'Transfer'], num_payments),
#     "ConfirmedBy": [fake.user_name() for _ in range(num_payments)],
#     "ReferenceNumber": [f"REF{random.randint(10000, 99999)}" for _ in range(num_payments)],
#     "Notes": [fake.sentence() for _ in range(num_payments)]
# })

# revenue_df = pd.DataFrame({
#     "RecognitionID": range(1, num_recognitions + 1),
#     "InvoiceID": np.random.choice(invoices["InvoiceID"], num_recognitions),
#     "DateRecognized": [datetime(2024, 1, 1) + timedelta(days=random.randint(0, 200)) for _ in range(num_recognitions)],
#     "Amount": np.round(np.random.uniform(50, 5000, num_recognitions), 2),
#     "Notes": [fake.bs() for _ in range(num_recognitions)]
# })

import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
from faker import Faker

fake = Faker()

# تعداد رکوردها
N_COUNTRIES       = 50
N_CUSTOMERS       = 300
N_BILLING_CYCLES  = 5
N_SERVICE_TYPES   = 300
N_TAXES           = 300
N_TARIFFS         = 300
N_CONTRACTS       = 300
N_INVOICES        = 300
N_PAYMENTS        = 300
N_RECOGNITIONS    = 300
N_INVOICE_LINES   = 1_000_000

# 1. Common.Country
countries = pd.DataFrame({
    "CountryID":   np.arange(1, N_COUNTRIES+1),
    "CountryName": [fake.country() for _ in range(N_COUNTRIES)],
    "CountryCode": [fake.country_code() for _ in range(N_COUNTRIES)]
})

# 2. Finance.Customer (FK → Country)
customers = pd.DataFrame({
    "CustomerID":    np.arange(1, N_CUSTOMERS+1),
    "CustomerCode":  [f"CUST{str(i).zfill(4)}" for i in range(1, N_CUSTOMERS+1)],
    "CustomerName":  [fake.company() for _ in range(N_CUSTOMERS)],
    "CustomerType":  np.random.choice(['Individual','Company','Foreign'], N_CUSTOMERS),
    "TIN":           [fake.bothify(text='??####??') for _ in range(N_CUSTOMERS)],
    "VATNumber":     [fake.bothify(text='VAT###??') for _ in range(N_CUSTOMERS)],
    "Phone":         [fake.phone_number() for _ in range(N_CUSTOMERS)],
    "Email":         [fake.company_email() for _ in range(N_CUSTOMERS)],
    "Address":       [fake.address().replace("\n",", ") for _ in range(N_CUSTOMERS)],
    "CountryID":     np.random.choice(countries["CountryID"], N_CUSTOMERS)
})

# 3. Finance.BillingCycle (سورس مستقل)
billing_cycles = pd.DataFrame({
    "BillingCycleID":    [i+1 for i in range(N_BILLING_CYCLES)],
    "CycleName":         ['Monthly','Quarterly','Annually','Bi-Weekly','Weekly'],
    "CycleLengthInDays": [30,90,365,14,7]
})

# 4. Finance.ServiceType
service_types = pd.DataFrame({
    "ServiceTypeID":   np.arange(1, N_SERVICE_TYPES+1),
    "ServiceName":     [fake.bs().title() for _ in range(N_SERVICE_TYPES)],
    "ServiceCategory": np.random.choice(['Unloading','Storage','Transport'], N_SERVICE_TYPES),
    "BaseRate":        np.round(np.random.uniform(50,500, N_SERVICE_TYPES),2),
    "UnitOfMeasure":   np.random.choice(['TEU','Hour','Ton'], N_SERVICE_TYPES),
    "Taxable":         np.random.choice([0,1], N_SERVICE_TYPES),
    "IsActive":        np.ones(N_SERVICE_TYPES, dtype=int)
})

# 5. Finance.Tax
taxes = pd.DataFrame({
    "TaxID":        np.arange(1, N_TAXES+1),
    "TaxName":      [f"Tax{str(i).zfill(3)}" for i in range(1, N_TAXES+1)],
    "TaxRate":      np.round(np.random.uniform(0.01,0.25, N_TAXES),2),
    "TaxType":      np.random.choice(['National','Service'], N_TAXES),
    "EffectiveFrom":[datetime(2020,1,1)+timedelta(days=i*30) for i in range(N_TAXES)],
    "EffectiveTo":  [None]*N_TAXES
})

# 6. Finance.Tariff (FK → ServiceType)
tariffs = pd.DataFrame({
    "TariffID":      np.arange(1, N_TARIFFS+1),
    "ServiceTypeID": np.random.choice(service_types["ServiceTypeID"], N_TARIFFS),
    "ValidFrom":     [datetime(2021,1,1)+timedelta(days=random.randint(0,1000)) 
                      for _ in range(N_TARIFFS)],
    "ValidTo":       [None]*N_TARIFFS,
    "UnitRate":      np.round(np.random.uniform(100,1000, N_TARIFFS),2)
})

# 7. Finance.Contract (FK → Customer, BillingCycle)
contracts = pd.DataFrame({
    "ContractID":     np.arange(1, N_CONTRACTS+1),
    "CustomerID":     np.random.choice(customers["CustomerID"], N_CONTRACTS),
    "ContractNumber": [f"CON{str(i).zfill(6)}" for i in range(1, N_CONTRACTS+1)],
    "StartDate":      [datetime(2023,1,1)+timedelta(days=random.randint(0,365)) 
                       for _ in range(N_CONTRACTS)],
    "EndDate":        [None if random.random()<0.2 else 
                       datetime(2024,1,1)+timedelta(days=random.randint(0,365)) 
                       for _ in range(N_CONTRACTS)],
    "BillingCycleID": np.random.choice(billing_cycles["BillingCycleID"], N_CONTRACTS),
    "PaymentTerms":   np.random.choice(['Net 30','Net 60','Prepaid'], N_CONTRACTS),
    "ContractStatus": np.random.choice(['Active','Expired'], N_CONTRACTS),
    "CreatedDate":    [datetime.now()-timedelta(days=random.randint(0,730)) 
                       for _ in range(N_CONTRACTS)]
})

# 8. Finance.Invoice (FK → Contract)
invoices = pd.DataFrame({
    "InvoiceID":     np.arange(1, N_INVOICES+1),
    "ContractID":    np.random.choice(contracts["ContractID"], N_INVOICES),
    "InvoiceNumber": [f"INV{str(i).zfill(7)}" for i in range(1, N_INVOICES+1)],
    "InvoiceDate":   [datetime(2024,1,1)+timedelta(days=random.randint(0,180)) 
                      for _ in range(N_INVOICES)],
    "DueDate":       [datetime(2024,1,1)+timedelta(days=random.randint(181,270)) 
                      for _ in range(N_INVOICES)],
    "Status":        np.random.choice(['Paid','Overdue','Cancelled'], N_INVOICES),
    "TotalAmount":   np.round(np.random.uniform(500,20000, N_INVOICES),2),
    "TaxAmount":     np.round(np.random.uniform(50,2000, N_INVOICES),2),
    "CreatedBy":     [fake.user_name() for _ in range(N_INVOICES)],
    "CreatedDate":   [datetime.now()-timedelta(days=random.randint(0,100)) 
                      for _ in range(N_INVOICES)]
})

# 9. Finance.Payment (FK → Invoice)
payments = pd.DataFrame({
    "PaymentID":       np.arange(1, N_PAYMENTS+1),
    "InvoiceID":       np.random.choice(invoices["InvoiceID"], N_PAYMENTS),
    "PaymentDate":     [datetime(2024,1,1)+timedelta(days=random.randint(0,200)) 
                        for _ in range(N_PAYMENTS)],
    "Amount":          np.round(np.random.uniform(50,10000, N_PAYMENTS),2),
    "PaymentMethod":   np.random.choice(['Cash','Card','Transfer'], N_PAYMENTS),
    "ConfirmedBy":     [fake.user_name() for _ in range(N_PAYMENTS)],
    "ReferenceNumber": [f"REF{random.randint(10000,99999)}" for _ in range(N_PAYMENTS)],
    "Notes":           [fake.sentence() for _ in range(N_PAYMENTS)]
})

#10. Finance.RevenueRecognition (FK → Invoice)
recognitions = pd.DataFrame({
    "RecognitionID":   np.arange(1, N_RECOGNITIONS+1),
    "InvoiceID":       np.random.choice(invoices["InvoiceID"], N_RECOGNITIONS),
    "DateRecognized":  [datetime(2024,1,1)+timedelta(days=random.randint(0,250)) 
                        for _ in range(N_RECOGNITIONS)],
    "Amount":          np.round(np.random.uniform(20,8000, N_RECOGNITIONS),2),
    "Notes":           [fake.bs() for _ in range(N_RECOGNITIONS)]
})

#11. Finance.InvoiceLine (FK → Invoice, ServiceType, Tax)
invoice_lines = pd.DataFrame({
    "InvoiceLineID":    np.arange(1, N_INVOICE_LINES+1),
    "InvoiceID":        np.random.choice(invoices["InvoiceID"], N_INVOICE_LINES),
    "ServiceTypeID":    np.random.choice(service_types["ServiceTypeID"], N_INVOICE_LINES),
    "TaxID":            np.random.choice(taxes["TaxID"], N_INVOICE_LINES),
    "Quantity":         np.random.randint(1,50, N_INVOICE_LINES),
    "UnitPrice":        np.round(np.random.uniform(20,500, N_INVOICE_LINES),2),
    "DiscountPercent":  np.round(np.random.uniform(0,0.3, N_INVOICE_LINES),2),
    "TaxAmount":        np.round(np.random.uniform(5,300, N_INVOICE_LINES),2),
    "NetAmount":        np.round(np.random.uniform(20,500, N_INVOICE_LINES),2)
})

# همه DataFrameها آماده‌اند و کلیدهای خارجی با منابع منطبق است.
# حالا می‌توانید با استفاده از تابعی مثل generate_insert_statements() خروجی SQL بسازید.




def generate_insert_statements(df, table_name):
    lines = []
    columns = ', '.join(df.columns)

    for _, row in df.iterrows():
        values = []
        for val in row:
            if pd.isnull(val):
                values.append("NULL")
            elif isinstance(val, str):
                # Escape single quotes by replacing ' with ''
                escaped = val.replace("'", "''")
                values.append("N'" + escaped + "'")
            elif isinstance(val, (datetime, pd.Timestamp)):
                values.append("'" + val.strftime("%Y-%m-%d") + "'")
            else:
                values.append(str(val))

        values_joined = ', '.join(values)
        sql = f"INSERT INTO {table_name} ({columns}) VALUES ({values_joined});"
        lines.append(sql)

    return lines


tables = {
    # 0) مرجع مشترک
    "Common.Country":          countries,
    # 1) دوره‌های مالی (بدون FK)
    "Finance.BillingCycle":    billing_cycles,
    # 2) تعرفه‌ها (FK→ServiceType)
    "Finance.ServiceType":     service_types,
    # 3) مالیات‌ها (بدون FK)
    "Finance.Tax":             taxes,
    # 4) تعرفه‌ی خدمات (FK→ServiceType)
    "Finance.Tariff":          tariffs,
    # 5) مشتریان (FK→Country)
    "Finance.Customer":        customers,
    # 6) قراردادها (FK→Customer, BillingCycle)
    "Finance.Contract":        contracts,
    # 7) فاکتورها (FK→Contract)
    "Finance.Invoice":         invoices,
    # 8) پرداخت‌ها (FK→Invoice)
    "Finance.Payment":         payments,
    # 9) شناسایی درآمد (FK→Invoice)
    "Finance.RevenueRecognition": recognitions,
    # 🔟 جزئیات فاکتور (FK→Invoice, ServiceType, Tax) — این بزرگ‌ترین جدول
    "Finance.InvoiceLine":     invoice_lines
}


sql_lines = []
for table_name, df in tables.items():
    sql_lines.extend(generate_insert_statements(df, table_name))

with open("insert_finance_data.sql", "w", encoding="utf-8") as f:
    f.write('\n'.join(sql_lines))

print("✅ File 'insert_finance_data.sql' created successfully.")
```

```python
import pandas as pd
import random
from faker import Faker
from datetime import datetime, timedelta
import os

# تنظیم Faker برای داده‌های انگلیسی
fake = Faker('en_US')

# تنظیمات تعداد داده‌ها
NUM_EMPLOYEES = 1000
NUM_ATTENDANCE = 1000000
NUM_SALARY_PAYMENTS = 400000

# مسیر ذخیره فایل‌های CSV
OUTPUT_DIR = './tradeportdb_data'
os.makedirs(OUTPUT_DIR, exist_ok=True)

# لیست مشاغل منطقی بندری
POSITIONS = ['Crane Operator', 'Forklift Driver', 'Dock Supervisor', 'Logistics Clerk', 'Port Manager', 
             'Maintenance Technician', 'Security Officer', 'Cargo Inspector']

# ----------------------------------
# تولید داده برای جدول Employee
# ----------------------------------
def generate_phone_number():
    """تولید شماره تلفن با فرمت +989XXXXXXXXX"""
    return f"+989{random.randint(100000000, 999999999)}"

def generate_national_id(used_ids):
    """تولید کد ملی 10 رقمی منحصربه‌فرد"""
    while True:
        nid = f"{random.randint(1000000000, 9999999999)}"
        if nid not in used_ids:
            used_ids.add(nid)
            return nid

employees = []
used_national_ids = set()
employment_status_weights = ['Active'] * 80 + ['OnLeave'] * 15 + ['Terminated'] * 5
gender_weights = ['Male'] * 70 + ['Female'] * 30
marital_status_weights = ['Married'] * 60 + ['Single'] * 40

for i in range(1, NUM_EMPLOYEES + 1):
    hire_date = fake.date_between(start_date=datetime(2015, 1, 1), end_date='today')
    birth_date = fake.date_between(start_date=datetime(1965, 1, 1), end_date=datetime(2007, 12, 31))
    employees.append({
        'EmployeeID': i,
        'FullName': fake.name(),
        'Position': random.choice(POSITIONS),
        'NationalID': generate_national_id(used_national_ids),
        'HireDate': hire_date,
        'BirthDate': birth_date,
        'Gender': random.choice(gender_weights),
        'MaritalStatus': random.choice(marital_status_weights),
        'Address': fake.address().replace('\n', ', '),
        'Phone': generate_phone_number(),
        'Email': fake.email(),
        'EmploymentStatus': random.choice(employment_status_weights)
    })

df_employees = pd.DataFrame(employees)
df_employees.to_csv(os.path.join(OUTPUT_DIR, 'employee.csv'), index=False, encoding='utf-8-sig')
print(f"Generated {len(df_employees)} records for Employee table.")

# ----------------------------------
# تولید داده برای جدول Attendance
# ----------------------------------
attendance = []
active_onleave_ids = df_employees[df_employees['EmploymentStatus'].isin(['Active', 'OnLeave'])]['EmployeeID'].tolist()
attendance_status_weights = ['Present'] * 85 + ['Late'] * 10 + ['Absent'] * 5

def is_workday(date):
    """بررسی اینکه تاریخ روز کاری است (دوشنبه تا شنبه)"""
    return date.weekday() < 6  # 0=دوشنبه، 5=شنبه، 6=یکشنبه

for i in range(1, NUM_ATTENDANCE + 1):
    # تولید تاریخ تصادفی، با احتمال بیشتر برای روزهای کاری
    while True:
        attendance_date = fake.date_between(start_date=datetime(2020, 1, 1), end_date='today')
        if is_workday(attendance_date) or random.random() < 0.1:  # 10% شانس برای یکشنبه
            break
    
    status = random.choice(attendance_status_weights)
    
    # زمان ورود و خروج فقط برای Present یا Late
    check_in_time = None
    check_out_time = None
    hours_worked = None
    if status in ['Present', 'Late']:
        check_in_hour = random.randint(7, 8) if status == 'Present' else random.randint(8, 9)
        check_in_time = datetime.strptime(f"{check_in_hour}:{random.randint(0, 59):02d}:00", '%H:%M:%S').strftime('%H:%M:%S')
        hours = random.uniform(6, 10)
        check_out_time = (datetime.strptime(check_in_time, '%H:%M:%S') + timedelta(hours=hours)).strftime('%H:%M:%S')
        hours_worked = round(hours, 2)
    
    attendance.append({
        'AttendanceID': i,
        'EmployeeID': random.choice(active_onleave_ids),
        'AttendanceDate': attendance_date,
        'Status': status,
        'CheckInTime': check_in_time if check_in_time else '00:00:00',
        'CheckOutTime': check_out_time if check_out_time else '00:00:00',
        'HoursWorked': hours_worked if hours_worked else 0.0,
        'Notes': fake.sentence()
    })

df_attendance = pd.DataFrame(attendance)
df_attendance.to_csv(os.path.join(OUTPUT_DIR, 'attendance.csv'), index=False, encoding='utf-8-sig')
print(f"Generated {len(df_attendance)} records for Attendance table.")

# ----------------------------------
# تولید داده برای جدول SalaryPayment
# ----------------------------------
salary_payments = []
payment_method_weights = ['BankTransfer'] * 90 + ['Cash'] * 10

for i in range(1, NUM_SALARY_PAYMENTS + 1):
    # تولید تاریخ پرداخت (معمولاً پایان ماه)
    payment_date = fake.date_between(start_date=datetime(2020, 1, 1), end_date='today')
    payment_date = payment_date.replace(day=28)  # فرض پرداخت در روز 28 هر ماه
    
    amount = round(random.uniform(2000, 6000), 2)
    bonus = round(random.uniform(200, 1000), 2)
    deductions = round(random.uniform(100, 500), 2)
    net_amount = amount + bonus - deductions
    
    salary_payments.append({
        'SalaryPaymentID': i,
        'EmployeeID': random.choice(active_onleave_ids),
        'PaymentDate': payment_date,
        'Amount': amount,
        'Bonus': bonus,
        'Deductions': deductions,
        'NetAmount': round(net_amount, 2),
        'PaymentMethod': random.choice(payment_method_weights),
        'ReferenceNumber': f"REF{random.randint(100000, 999999)}"
    })

df_salary_payments = pd.DataFrame(salary_payments)
df_salary_payments.to_csv(os.path.join(OUTPUT_DIR, 'salary_payment.csv'), index=False, encoding='utf-8-sig')
print(f"Generated {len(df_salary_payments)} records for SalaryPayment table.")
```
